#MNIST com Rede Convolucional (Convnet)

```python
inputs = keras.Input(shape=(28, 28, 1))
x = keras.layers.Conv2D(filters=32, kernel_size=3, activation="relu")(inputs)
x = keras.layers.MaxPooling2D(pool_size=2)(x)
x = keras.layers.Conv2D(filters=64, kernel_size=3, activation="relu")(x)
x = keras.layers.MaxPooling2D(pool_size=2)(x)
x = keras.layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x)
x = keras.layers.Flatten()(x)
outputs = keras.layers.Dense(10, activation="softmax")(x)
modelo = keras.Model(inputs=inputs, outputs=outputs)
```

```python
modelo.summary()
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images=train_images.reshape((60000, 28, 28, 1))
train_images=train_images.astype("float32")/255

test_images=test_images.reshape((10000, 28, 28, 1))
test_images=test_images.astype("float32")/255
```

```python
modelo.compile(
    optimizer="rmsprop",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
historico = modelo.fit(train_images, train_labels, epochs=10, batch_size=128)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(historico.history['accuracy'])
plt.title('Acurcia por poca')
plt.xlabel('poca')
plt.ylabel('acurcia')
plt.legend(['treino', 'validao'])

plt.subplot(1, 2, 2)
plt.plot(historico.history['loss'])
plt.title('Perda por poca')
plt.xlabel('poca')
plt.ylabel('perda')
plt.legend(['treino', 'validao'])
plt.show()
```

```python
test_loss, test_acc = modelo.evaluate(test_images, test_labels)
print(test_acc)
```

```python
from keras.preprocessing.image import ImageDataGenerator
#image augmentation
train_datagen =ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

training_set = train_datagen.flow_from_directory('train',
                                                 target_size=(64,64),
                                                 batch_size=32,
                                                 class_mode='binary')
```

#Criando a Convnet (CNN)

```python
import tensorflow as tf

```

```python
modelo = tf.keras.models.Sequential()
modelo.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64,64,3]))
modelo.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
modelo.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
modelo.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
modelo.add(tf.keras.layers.Flatten())
modelo.add(tf.keras.layers.Dense(units=128, activation='relu'))
modelo.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
modelo.compile(
    optimizer="rmsprop",
    loss="binary_crossentropy",
    metrics=['accuracy']
)
```

```python
historico = modelo.fit(x=training_set, validation_data=val_set, epochs=10)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(historico.history['accuracy'])
plt.plot(historico.history['val_accuracy'])
plt.title('Acurcia por poca')
plt.xlabel('poca')
plt.ylabel('acurcia')
plt.legend(['treino', 'validao'])

plt.subplot(1, 2, 2)
plt.plot(historico.history['loss'])
plt.plot(historico.history['val_loss'])
plt.title('Perda por poca')
plt.xlabel('poca')
plt.ylabel('perda')
plt.legend(['treino', 'validao'])
plt.show()
```
